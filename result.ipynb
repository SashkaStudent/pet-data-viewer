# Дискретное вейвлет преобразование

Были взяты данные магнитной обсерватории "Паратунка" за 2017 год. 
<br>Выборка данных основывалась на K-index и распределялись следующим образом:
|       Значение      |      Состояние         |
|--------------|------------------------|
| K≤=2 и менее | спокойное              |
| K=2…3        | слабо возмущенное      |
| K=4          | возмущенное            |
| K=5…6        | магнитная буря         |
| K≥=7 и более | большая магнитная буря |
